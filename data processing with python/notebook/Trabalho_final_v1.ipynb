** 3) Question ** Distribution of good and bad payers per state

```python
# Data Filter
estado_counts = pd.DataFrame({'Estado': ['SP', 'RJ', 'MG', 'BA', 'RS', 'PR', 'SC', 'PE', 'CE', 'GO', 'DF', 'ES', 'MA', 'PB', 'RN', 'MT', 'AL', 'PI', 'MS', 'PA', 'SE', 'AM', 'RO', 'TO', 'AP', 'RR', 'AC'],
                              'Bons_Pagadores': [5000, 4000, 3800, 3200, 3000, 2500, 2400, 2000, 1800, 1500, 1400, 1300, 1200, 1100, 1000, 900, 800, 700, 600, 500, 400, 300, 200, 150, 100, 80, 50],
                              'Maus_Pagadores': [3000, 2500, 2000, 1800, 1700, 1500, 1400, 1300, 1200, 1100, 900, 850, 800, 700, 650, 600, 550, 500, 450, 400, 350, 300, 250, 200, 150, 100, 50]})

# Normalizing the values ​​for the proportion of bad payers per state
estado_counts_norm = estado_counts.set_index('Estado').div(estado_counts.set_index('Estado').sum(axis=1), axis=0)

# Order States by the proportion of bad payers
estado_counts_sorted = estado_counts_norm.sort_values(by='Maus_Pagadores', ascending=False)
estado_counts_sorted.head(n=5)
```

The code aims to calculate and view the proportion of bad payers per state and identify which states have the highest default rate. Next, we detail each step, create a fictitious dataframe containing three state columns: Brazilian state acronym, good_pators, number of customers who are not defaulting, maus_pators amount of customers who are delinquent. The code converts absolute values ​​(number of customers) into proportions, ie, transforms the numbers into percentages within each state. This is done as follows. Defines the state as an index to facilitate data manipulation, we turn the state column into the Dataframe index. Divides each value by the total of customers in the state. For each state, the amount of good payers and bad payers is divided by the total of customers in the state, the amounts now represent the relative proportion within each state (and no more absolute numbers).

```python
# Bar Chart
fig, ax = plt.subplots(figsize=(20.5, 10))
estado_counts_sorted.plot(kind="barh", stacked=True, colormap="coolwarm", ax=ax, alpha=0.85)
plt.title("Proporção de Bons e Maus Pagadores por Estado Residencial", fontsize=14, fontweight="bold")
plt.xlabel("Proporção de Clientes", fontsize=12)
plt.ylabel("Estado Residencial", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(["Bons Pagadores", "Maus Pagadores"], title="Classificação", fontsize=11)

# Percentage labels in the bars
for idx, (estado, valores) in enumerate(estado_counts_sorted.iterrows()):
    pct_mau = valores['Maus_Pagadores']  # Proporção de maus pagadores
    plt.text(valores['Bons_Pagadores'] + 0.02, idx, f"{pct_mau:.0%}", va="center", fontsize=10, color="black", fontweight="bold")

# Inverter axis y for better view
plt.gca().invert_yaxis()

# Viewing Graph
plt.show()
```

```python
# Select only numerical columns
df_numerical = data.select_dtypes(include=['int64', 'float64'])

# Outlier Detection Using the IQR (Interquartile Range) method
outliers_dict = {}

for col in df_numerical.columns:
    Q1 = np.percentile(df_numerical[col], 25)  # Primeiro quartil
    Q3 = np.percentile(df_numerical[col], 75)  # Terceiro quartil
    IQR = Q3 - Q1  # Intervalo interquartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_numerical[(df_numerical[col] < lower_bound) | (df_numerical[col] > upper_bound)]
    
# Stores the outliers count by variable
    outliers_dict[col] = len(outliers)

# Display how many outliers were found in each numerical variable
print("\n Contagem de Outliers por Variável:")
print()
for key, value in outliers_dict.items():
    print()
    print(f"{key}: {value} outliers")
```

** Remove extreme outliers (IQR method) **

```python
# Create a copy of Dataframe to avoid direct changes
df_cleaned = data.copy()

# List of columns with identified outliers
outlier_columns = ["DIA_PAGAMENTO", "TIPO_ENDERECO_POSTAL", "ESTADO_CIVIL", "QUANT_DEPENDENTES",
                   "NACIONALIDADE", "TIPO_RESIDENCIA", "MESES_RESIDENCIA", "RENDA_PESSOAL_MENSAL",
                   "OUTRAS_RENDAS", "FLAG_VISA", "FLAG_MASTERCARD", "FLAG_DINERS",
                   "FLAG_AMERICAN_EXPRESS", "FLAG_OUTROS_CARTOES", "QUANT_CONTAS_BANCARIAS",
                   "QUANT_CONTAS_BANCARIAS_ESPECIAIS", "VALOR_PATRIMONIO_PESSOAL",
                   "QUANT_CARROS", "MESES_NO_TRABALHO", "CODIGO_PROFISSAO", "TIPO_OCUPACAO",
                   "NIVEL_EDUCACIONAL_CONJUGE", "PRODUTO", "IDADE"]

# Apply IQR method to remove outliers
for col in outlier_columns:
    Q1 = np.percentile(df_cleaned[col], 25)
    Q3 = np.percentile(df_cleaned[col], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

# Check how many data were removed
print(f"\n Dados antes da remoção: {data.shape[0]}")
print(f" Dados depois da remoção: {df_cleaned.shape[0]}")
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import RocCurveDisplay

# Check if the problem is binary or multiclass
num_classes = len(np.unique(y_train))

# Define the models
models = {"Naive Bayes": GaussianNB(),
          "Decision Tree": DecisionTreeClassifier(random_state=42),
          "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
          "Logistic Regression": LogisticRegression(random_state=50, max_iter=1000)}

# If it's multiclass, binarize the target
if num_classes > 2:
    from sklearn.preprocessing import label_binarize
    y_train_bin = label_binarize(y_train, classes=np.unique(y_train))
    y_test_bin = label_binarize(y_test, classes=np.unique(y_train))
else:
    y_train_bin = y_train
    y_test_bin = y_test

# Train and plot Roc to each model
for model_name, model in models.items():
    print(f"Training {model_name}...")
    
    classifier = OneVsRestClassifier(model) if num_classes > 2 else model
    classifier.fit(X_train, y_train_bin)

# Probabilities of predictions
    y_prob = classifier.predict_proba(X_test)

# Check if there are nans in probabilities
    if np.isnan(y_prob).sum() > 0:
        print(f"There are NaN values in the predictions for model {model_name}. Skipping this model.")
        continue

# Create the ROC
    plt.figure(figsize=(10, 8))

    if num_classes > 2:
        for i in range(y_test_bin.shape[1]):  # Para cada classe
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f"Class {i} (AUC = {roc_auc:.2f})")
    else:
        fpr, tpr, _ = roc_curve(y_test_bin, y_prob[:, 1])  # Apenas a classe positiva
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")

# Finish graph
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Linha diagonal (random guessing)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve for {model_name}")
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
```
